```python
import requests
import json
import os

# Specify the folder path where you want to save the JSON files
folder = "C:/Users/lmhmo/indexer/blocks"  # Adjust this path as needed

# Define the range of block heights you want to retrieve (e.g., from 1 to 100)
start_height = 1 #specify block height you want to start with
end_height = 100 # specify block height you want to end with

for height in range(start_height, end_height + 1):
    # Make an API request for each block
    response = requests.get(f"http://116.202.143.93:1317/cosmos/base/tendermint/v1beta1/blocks/{height}")

    # Construct the filename based on the block height
    filename = os.path.join(folder, f"{height}.json")

    # Save the response content to the dynamically named file
    with open(filename, "wb") as f:
        f.write(response.content)

    print(f"Block {height} data saved to {filename}")

print("Data retrieval complete for all blocks.")
```
