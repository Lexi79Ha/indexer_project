```python
import json
import base64

def decode_and_count_characters(json_file_path):
    try:
        with open(json_file_path, "r") as file:
            data = json.load(file)
            txs_base64 = data.get("txs", "")  # Get the value of "txs" or an empty string if not present

            # Decode the Base64 string
            txs_bytes = base64.b64decode(txs_base64)
            txs_decoded = txs_bytes.decode("utf-8")

            # Count the characters
            total_characters = len(txs_decoded)
            print(f"Total characters in 'txs' (after Base64 decoding): {total_characters}")
    except FileNotFoundError:
        print(f"File '{json_file_path}' not found.")
    except json.JSONDecodeError:
        print(f"Error decoding JSON data in '{json_file_path}'.")

# Usage example:
json_file_path = "C:/Users/lmhmo/indexer/blocks/3102393.json"
decode_and_count_characters(json_file_path)

```

```python
import json
import base64

def decode_and_count_characters(json_file_path):
    try:
        with open(json_file_path, "r") as file:
            data = json.load(file)
            txs_list = data.get("block", {}).get("data", {}).get("txs", [])  # Get the list of transactions

            total_characters = 0
            for txs_base64 in txs_list:
                # Decode the Base64 string using "latin-1" encoding
                txs_bytes = base64.b64decode(txs_base64)
                txs_decoded = txs_bytes.decode("latin-1")
                total_characters += len(txs_decoded)

            print(f"Total characters in all transactions (after Base64 decoding): {total_characters}")
    except FileNotFoundError:
        print(f"File '{json_file_path}' not found.")
    except json.JSONDecodeError:
        print(f"Error decoding JSON data in '{json_file_path}'.")

# Usage example:
json_file_path = "C:/Users/lmhmo/indexer/blocks/3102393.json"
decode_and_count_characters(json_file_path)
```

```python
import json
import base64

def decode_and_print_characters(json_file_path):
    try:
        with open(json_file_path, "r") as file:
            data = json.load(file)
            txs_list = data.get("block", {}).get("data", {}).get("txs", [])  # Get the list of transactions

            total_characters = 0
            for txs_base64 in txs_list:
                # Decode the Base64 string using "latin-1" encoding
                txs_bytes = base64.b64decode(txs_base64)
                txs_decoded = txs_bytes.decode("latin-1")
                total_characters += len(txs_decoded)

                # Print the decoded transaction
                print(f"Decoded transaction: {txs_decoded}")

            print(f"Total characters in all transactions (after Base64 decoding): {total_characters}")
    except FileNotFoundError:
        print(f"File '{json_file_path}' not found.")
    except json.JSONDecodeError:
        print(f"Error decoding JSON data in '{json_file_path}'.")

# Usage example:
json_file_path = "C:/Users/lmhmo/indexer/blocks/3102334.json"
decode_and_print_characters(json_file_path)
```
